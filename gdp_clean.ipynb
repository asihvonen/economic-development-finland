```python
import pandas as pd
import numpy as np


df = pd.read_excel('gdp.xlsx', index_col=False)

metrics = df['Unnamed: 0']

df = df.drop('Unnamed: 0', axis=1)
df

N = 6

dfs = np.split(df, np.arange(N, len(df.columns), N), axis=1)

dfs = dfs[1:]

dfs[1]
```

```python
first_columns = [df.iloc[:, 0] for df in dfs]

new_df = pd.concat(first_columns, axis=1)

new_df.index = metrics

new_df = new_df.drop(new_df.index[0])

new_df = new_df.T

new_df
```
