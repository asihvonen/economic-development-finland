```python
# methods for k means clustering

from sklearn.cluster import KMeans
import numpy as np
import matplotlib.pyplot as plt
from sklearn.metrics import silhouette_score

def kmeans_clustering(data: pd.DataFrame, n_clusters: int):
    kmeans = KMeans(n_clusters=n_clusters, random_state=42)
    kmeans.fit(data)

    data["Cluster"] = kmeans.labels_
    return data, kmeans

def elbow_method(data: pd.DataFrame, max_k: int):
    wcss = []  # within-cluster sum of squares
    ks = list(range(1, max_k + 1))
    for k in ks:
        kmeans = KMeans(n_clusters=k, random_state=42)
        kmeans.fit(data)
        wcss.append(kmeans.inertia_)

    plt.plot(ks, wcss)
    plt.xlabel("Number of clusters")
    plt.ylabel("WCSS")
    plt.title("Elbow Method For Optimal k")
    plt.xticks(ks)  # ensure integer ticks
    plt.show()

def silhouette_method(data: pd.DataFrame, max_k: int):
    ks = list(range(2, max_k + 1))
    silhouette_scores = []
    for k in ks:
        kmeans = KMeans(n_clusters=k, random_state=42)
        kmeans.fit(data)
        score = silhouette_score(data, kmeans.labels_)
        silhouette_scores.append(score)

    plt.plot(ks, silhouette_scores)
    plt.xlabel("Number of clusters")
    plt.ylabel("Silhouette Score")
    plt.title("Silhouette Method For Optimal k")
    plt.xticks(ks)  # ensure integer ticks
    plt.show()
```

### Clustered by each year over indices

```python
indices = ["Economic prosperity", "Business environment", "Demographic sustainability"]
```

```python
# for each year, cluster regions based on indices, setting 5 clusters

clusters_df = pd.DataFrame(columns=[])

for year in indices_region_df["Year"].unique(): 
    year_data = indices_region_df[indices_region_df["Year"] == year]
    year_data_scaled = scaler.fit_transform(year_data.drop(columns=["Region", "Year"]))
    year_data_scaled_df = pd.DataFrame(year_data_scaled, columns=year_data.drop(columns=["Region", "Year"]).columns)
    
    clustered_year_data, kmeans_model = kmeans_clustering(year_data_scaled_df, n_clusters=5)
    clustered_year_data["Region"] = year_data["Region"].values
    clustered_year_data["Year"] = year
    
    # visualize clusters in PCA space
    pca = PCA(n_components=2)
    pca_result = pca.fit_transform(year_data_scaled_df)
    year_data_scaled_df["PCA1"] = pca_result[:, 0]
    year_data_scaled_df["PCA2"] = pca_result[:, 1]
    year_data_scaled_df["Cluster"] = clustered_year_data["Cluster"].values  # ensure cluster info is present

    # make dataframe with years as columns, region as row, and cluster as value, updating clusters_df with each year
    cluster_pivot = year_data_scaled_df.pivot_table(index="Region", columns="Year", values="Cluster")
    clusters_df = pd.concat([clusters_df, cluster_pivot], axis=1)

    # Add Region and Year from year_data for hover info
    year_data_scaled_df["Region"] = year_data["Region"].values
    year_data_scaled_df["Year"] = year

    fig = px.scatter(
        year_data_scaled_df,
        x="PCA1",
        y="PCA2",
        color="Cluster",
        hover_data=["PCA1", "PCA2", "Region", "Year"],
        title=f"K-Means Clustering Results for {year}"
    )
    fig.show()
```
